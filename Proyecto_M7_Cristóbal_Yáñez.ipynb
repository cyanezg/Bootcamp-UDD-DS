# Análisis Exploratorio del Dataset "Your Career Aspirations of GenZ"

## 1. Resumen General del Dataset

Comenzamos cargando el dataset y examinando su estructura básica: número de filas y columnas, tipos de datos, y verificando valores nulos o duplicados.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el dataset
df = pd.read_csv("/content/drive/MyDrive/bootcamp/data_m7/Your Career Aspirations of GenZ.csv")

# Mostrar número de filas y columnas
print("Dimensiones (filas, columnas):", df.shape)

# Mostrar las primeras filas para ver la estructura
display(df.head(5))

# Información general de columnas y tipos
print("\nInformación del DataFrame:")
print(df.info())

# Contar valores nulos en cada columna
print("\nValores nulos por columna:")
print(df.isnull().sum())

# Contar filas duplicadas
print("\nCantidad de filas duplicadas:")
print(df.duplicated().sum())
```

**Observaciones generales:** El dataset tiene **235 filas y 15 columnas**, lo que indica 235 encuestas completadas con 15 preguntas/respuestas por participante. De acuerdo al `info()`, la mayoría de las columnas son de tipo `object` (categóricas representadas como texto), salvo quizás una o dos columnas numéricas. En particular, la columna de código postal (“Zip Code / Pin Code”) es numérica (int) y otra columna correspondiente a una escala de 1 a 10 (relacionada con la misión con impacto social, que veremos luego) también aparece como numérica.

No se encontraron valores nulos en ninguna columna – cada pregunta fue respondida por todos los encuestados (los conteos `non-null` son 235 en todas las columnas). Tampoco hay filas duplicadas (el resultado de `df.duplicated().sum()` es 0), por lo que no hace falta eliminar registros repetidos. En resumen, **el dataset está completo y sin datos faltantes ni duplicados**. Esto sugiere que la calidad de los datos es buena de entrada y podemos proceder con el análisis exploratorio.

Las columnas del dataset representan las siguientes informaciones recopiladas en la encuesta:

- **Your Current Country**: País actual del encuestado.  
- **Your Current Zip Code / Pin Code**: Código postal (en su mayoría de India, dado el contexto de la encuesta).  
- **Your Gender**: Género del encuestado (por los valores únicos, parece haber solo “Male” y “Female”).  
- **Which of the below factors influence the most about your career aspirations?**: Factor que más influye en sus aspiraciones profesionales (opciones categóricas como familia, redes sociales, etc.).  
- **Would you definitely pursue a Higher Education / Post Graduation outside of India? If only you have to self sponsor it.**: Si perseguiría estudios de posgrado fuera de India, asumiendo que debe autofinanciarlos (básicamente sí o no, con alguna condición).

- **What is the most preferred working environment for you.**: Ambiente de trabajo preferido (remoto, híbrido, oficina, etc.).  
- **Which of the below Employers would you work with.**: Tipo de empleador con el que le gustaría trabajar (por las respuestas, parecen ser opciones descriptivas, e.j. “empleador que recompensa el aprendizaje”, “que impulsa tus límites”, etc.).  
- **Which type of learning environment that you are most likely to work in?**: Tipo de entorno de aprendizaje preferido para desarrollarse profesionalmente (ej. aprendizaje autodidacta, con instructor, aprender observando, etc. – esta pregunta permitía seleccionar múltiples opciones).

- **What type of Manager would you work without looking into your watch?**: Tipo de gerente/líder bajo el cual trabajaría “sin ver el reloj” (es decir, con el que se sentiría a gusto trabajando largas horas).  
- **Which of the following setup you would like to work?**: Tipo de entorno/estructura en la que le gustaría trabajar (ej. startup, empresa grande, sector público, etc.).

### Factores que influyen en las aspiraciones profesionales

La siguiente pregunta identifica **el factor que más influye** en las aspiraciones de carrera de los jóvenes. Las opciones incluyen: la influencia de padres, personas que han cambiado el mundo, redes sociales profesionales, influencers con carreras exitosas, o personas de su círculo cercano (amigos/mentores) excluyendo familia.

- **“My Parents”** (Mis padres) es **otro factor muy mencionado**, evidenciando que la **familia inmediata** sigue jugando un papel fundamental en las decisiones de carrera para una gran parte de los Gen Z. - **“People from my circle, but not family members”** (personas de mi círculo cercano, pero no familiares) y **“Influencers who had successful careers”** (influencers o figuras públicas con carreras exitosas) fueron seleccionados por un grupo más reducido de encuestados. Esto sugiere que, comparativamente, **Gen Z confía más en ejemplos de gran impacto (mundial o familiar) que en pares o celebridades de internet** a la hora de formarse aspiraciones profesionales.

En síntesis, hay una mezcla de **influencias tradicionales (familiares)** y **modernas (redes sociales)**, pero los **modelos a seguir inspiradores** —sean héroes globales o padres— parecen tener la mayor influencia.

### Aspiración de Educación de Posgrado en el Extranjero

La pregunta: *“¿Definitivamente cursarías una Educación Superior/Posgrado fuera de India si tuvieras que autofinanciarte?”* busca entender la **disposición de los jóvenes a estudiar en el extranjero con sus propios medios**. Las posibles respuestas son:
- **“Yes, I will earn and do that”** – Sí, lo haría (estoy dispuesto a ganar/ahorrar dinero para costearlo).
- **“No, but if someone could bear the cost I will”** – No con mi dinero, pero sí lo haría si alguien más cubre el costo (ej. beca, padres, etc.).
- **“No, I would not be pursuing Higher Education outside of India”** – No, no lo haría bajo ninguna circunstancia si implica salir de India.

Veamos la distribución de estas respuestas:

```python
edu_counts = df['Would you definitely pursue a Higher Education / Post Graduation outside of India ? If only you have to self sponsor it.'].value_counts()
display(edu_counts)

plt.figure(figsize=(7,4))
edu_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title("¿Realizarías un posgrado fuera de India autofinanciado?")
plt.ylabel("")  # oculta la etiqueta del eje y
plt.show()
```

**Observación:** Encontramos **tres grupos** de casi todos los tamaños, mostrando las diferentes actitudes hacia cursar posgrados en el extranjero:

- **La mayoría relativa** de los encuestados (~alrededor de la mitad) optó por la respuesta condicional: *“No, pero si alguien costea, sí lo haría”*. Es decir, **muchos Gen Z estarían dispuestos a estudiar fuera de India solo si no tienen que pagarlo de su bolsillo** (por ejemplo, mediante becas o apoyo financiero externo). Esto refleja un enfoque pragmático: existe interés en la educación internacional, pero el costo es un impedimento significativo para muchos.
- Un grupo considerable (aunque menor que el anterior) respondió *“Sí, ahorraré/ganaré dinero y lo haré”*. Este grupo, quizás alrededor de un tercio de los encuestados, **está decidido a perseguir estudios de posgrado en el extranjero incluso si implica autofinanciarse**. Indica un alto nivel de motivación y compromiso con la idea de obtener una educación global, asumiendo sacrificios económicos.
- Una **minoría** de encuestados (aprox. 10-20%) dijo *“No lo haría”* en ningún caso. Este grupo **no planea realizar estudios fuera de India**, ya sea por preferencia personal de quedarse en el país, falta de interés en educación de posgrado, o porque consideran inviable irse fuera incluso con financiamiento.

En resumen, **la mayoría de los Gen Z indios sí aspira a tener educación internacional** de posgrado, pero **principalmente dependerían de apoyo financiero para hacerlo**, mientras que un subconjunto más comprometido está dispuesto a costearlo por su cuenta.

### Preferencia de Estabilidad Laboral (¿Trabajar ≥3 años con el mismo empleador?)

La pregunta sobre estabilidad: *“¿Qué tan probable es que trabajes para un solo empleador por 3 años o más?”* refleja la actitud hacia la permanencia en un trabajo (vs. movilidad laboral). Las opciones son:
- **“Will work for 3 years or more”** – Sí trabajaría 3 años o más con un empleador (alta estabilidad).
- **“This will be hard to do, but if it is the right company I would try”** – Sería difícil, pero si es la compañía adecuada lo intentaría (dependiente de circunstancias, estabilidad condicional).
- **“No way, 3 years with one employer is crazy”** – Ni pensarlo; 3 años con el mismo empleador es una locura (muy baja estabilidad, propenso a cambio frecuente).

Veamos la distribución:


```python
loyalty_counts = df['How likely is that you will work for one employer for 3 years or more ?'].value_counts()
display(loyalty_counts)

plt.figure(figsize=(8,4))
loyalty_counts.plot(kind='bar', color="orange")
plt.title("¿Trabajarías 3 años o más en el mismo empleo?")
plt.ylabel("Cantidad de encuestados")
plt.xticks(rotation=0)
plt.show()
```

**Observación:** Las actitudes hacia la **lealtad laboral** muestran una clara preferencia hacia la flexibilidad, aunque con matices:
- La mayoría de los encuestados eligió la opción intermedia: *“Sería difícil, pero si es la empresa correcta, lo intentaría”*. Es decir, **muchos Gen Z no se ven permaneciendo largos periodos por defecto, a menos que encuentren un trabajo realmente satisfactorio o acorde a sus expectativas**. Esta respuesta matizada sugiere que valoran las buenas condiciones/cultura laboral; si las tienen, estarían dispuestos a ser más estables.
- Un segundo grupo significativo respondió *“Trabajaría 3 años o más”* sin problemas. Este grupo (aunque más pequeño que el anterior) indica una **mentalidad más tradicional o estable**, con disposición a comprometerse a largo plazo con un empleador. Representa a quienes quizás priorizan la estabilidad o visualizan crecimiento dentro de una misma organización.
- Una minoría notable dijo *“Ni loco 3 años con uno”*. Estos **rechazan firmemente la idea de permanecer tanto tiempo** en el mismo lugar, reflejando una mentalidad muy inquieta o enfocada en cambiar de oportunidades frecuentemente (rasgo a veces atribuido a los Gen Z). Afortunadamente no es la mayoría, pero sí un segmento que destaca la preferencia por la **movilidad laboral y la diversidad de experiencias** en poco tiempo.

En general, **la mayoría sería flexible pero selectiva**: *estarían más de 3 años solo si vale la pena*. Unos pocos son totalmente **"job hoppers"** y otros más **leales**, pero condicionados a que la empresa sea la adecuada.

### ¿Trabajarías en una empresa sin misión claramente definida?

Esta pregunta es de **sí/no**: *“¿Trabajarías para una empresa cuya misión no esté claramente definida y pública?”*. Mide la importancia de la **transparencia de la misión empresarial** para los jóvenes.

Solo una pequeña fracción dijo que *sí trabajaría* en una empresa así, lo cual sugiere que **pocos están dispuestos a entrar a ciegas a una organización sin saber bien sus valores u objetivos públicos**.

En resumen, **la transparencia y comunicación de la misión corporativa es casi un requisito para esta generación al elegir empleador**. Esto es consistente con la idea de que Gen Z busca alineación de valores y propósito en el lugar de trabajo.

### ¿Trabajarías en una empresa con misión desalineada de sus acciones/producto?

Otra pregunta dicotómica: *“¿Qué tan probable es que trabajes para una empresa cuya misión esté desalineada con sus acciones públicas o incluso con su producto?”*. Las respuestas posibles son esencialmente **"Sí trabajaría"** vs **"No trabajaría"** (en los datos aparecen como “Will work for them” y “Will NOT work for them”).

```python
mission_align_counts = df['How likely would you work for a company whose mission is misaligned with their public actions or even their product ?'].value_counts()
display(mission_align_counts)

plt.figure(figsize=(5,4))
mission_align_counts.plot(kind='bar', color=["#99ff99","#ff9999"])
plt.title("¿Trabajarías en una empresa con misión desalineada?")
plt.ylabel("Número de encuestados")
plt.xticks(rotation=0)
plt.show()
```

En general, **Gen Z muestra una postura firme de no tolerar la falta de integridad en las empresas**: quieren que la misión declarada de la compañía se refleje realmente en sus prácticas y productos, de lo contrario preferirán no ser parte de ella.

### Disposición a trabajar en una empresa sin impacto social

Esta pregunta se midió en **escala del 1 al 10**: *“¿Qué tan dispuesto estarías a trabajar para una empresa cuya misión no trae impacto social?”*. Un 1 indicaría “nada dispuesto” y 10 “muy dispuesto”, asumiendo la escala lineal. En los datos, vemos valores del 1 al 10.

Analicemos la distribución de esas calificaciones:

```python
impact_ratings = df['How likely would you work for a company whose mission is not bringing social impact ?']
plt.figure(figsize=(6,4))
impact_ratings.plot(kind='hist', bins=10, color="purple", edgecolor="black")
plt.title("Distribución de puntuaciones: trabajar en empresa SIN impacto social")
plt.xlabel("Calificación (1 = nada dispuesto, 10 = muy dispuesto)")
plt.ylabel("Número de encuestados")
plt.show()

print("Valor mínimo:", impact_ratings.min(), " | Valor máximo:", impact_ratings.max())
print("Media:", impact_ratings.mean(), " | Mediana:", impact_ratings.median())
```

**Observación:** Las calificaciones van de **1 a 10**, cubriendo todo el rango posible, lo que indica variabilidad en las opiniones. Sin embargo, la distribución **está sesgada hacia valores bajos**. Muchos encuestados dieron puntuaciones **bajas (cercanas a 1)**, reflejando que **la mayoría no está dispuesta a trabajar en una empresa sin impacto social significativo**. Es decir, un buen número de jóvenes prácticamente rechazaría ese tipo de empresa (valorando mucho el impacto social en su trabajo).

También hay respuestas en el rango medio y alto, lo cual significa que **algunos Gen Z sí estarían algo dispuestos a trabajar en una empresa sin impacto social**. Tal vez estos encuestados priorizan otros factores (como crecimiento profesional, salario, tecnología, etc.) por encima del impacto social de la compañía. ### Entorno de aprendizaje preferido

La pregunta: *“¿Qué tipo de entorno de aprendizaje es en el que es más probable que trabajes?”* busca conocer **cómo prefieren los Gen Z aprender o desarrollarse en su trabajo**. Esta pregunta permitía **selección múltiple**, y las opciones incluyen métodos como:
- **Instructor or Expert-led Learning Programs** – Aprendizaje guiado por un instructor o experto (cursos, trainings formales).
- **Self Paced Learning Portals** – Aprendizaje autodirigido en portales online a tu propio ritmo.
- **Learning by observation** – Aprender mediante observación (ej. shadowing, mirando a otros).
- **Trial and error** – Aprender mediante prueba y error (experimentos, proyectos propios).
- **Other learning environment** – Otros.

Para analizar las preferencias, conviene contar **cuántas veces** fue mencionada cada modalidad (teniendo en cuenta que cada persona pudo elegir más de una). Extraeremos las opciones seleccionadas por los encuestados:


```python
# Separar las opciones múltiples de la columna de entorno de aprendizaje
learning_series = df['Which type of learning environment that you are most likely to work in ?'].str.split(',', expand=True).stack()
learning_series = learning_series.str.strip()  # quitar espacios en extremos de cada opción
learning_counts = learning_series.value_counts()
display(learning_counts)

plt.figure(figsize=(8,4))
learning_counts.plot(kind='bar', color="mediumpurple")
plt.title("Preferencias de entorno/metodología de aprendizaje")
plt.ylabel("Veces seleccionada (total menciones)")
plt.xticks(rotation=45, ha='right')
plt.show()
```

**Observación:**  Las preferencias de entorno de aprendizaje de Gen Z muestran una **preferencia clara por el aprendizaje guiado por expertos**:
- La opción **“Instructor or Expert-led”** fue la más común. Una gran proporción de jóvenes valora aprender de instructores calificados o programas estructurados de formación. Aproximadamente dos tercios de los encuestados incluyeron esta opción, indicando que **el aprendizaje guiado (por ejemplo, cursos, talleres, mentoring) es visto como muy eficaz**.
- En segundo lugar, pero muy por detrás, aparece el **aprendizaje autodidacta a tu propio ritmo (“Self Paced Learning Portals”)**. Alrededor de una cuarta parte de los participantes seleccionó esta opción. Esto sugiere que **si bien muchos aprecian la flexibilidad de plataformas online para aprender a su ritmo, no es la estrategia principal para la mayoría**, al menos no en exclusividad.
- Otras formas, como **“Learning by observation”** y **“Trial and error”**, fueron mencionadas por algunos pero en proporciones bastante menores. Pocos confían en aprender solo observando o mediante pura experimentación sin guía; la **Gen Z prefiere combinar esos métodos con otros más formales**. De hecho, varios de los encuestados que eligieron "trial and error" o "observation" también combinaron estas con métodos guiados o autodidactas (muchos seleccionaron más de una opción).
- *Other* no fue común casi, lo que indica que las opciones proporcionadas cubrían prácticamente todas las preferencias típicas.

En conclusión, **Gen Z prefiere un enfoque de aprendizaje estructurado y mentorizado**, complementado en menor medida por recursos autodirigidos. Quieren aprender en el trabajo mediante **programas con instructores/mentores**, aunque algunos también aprovechan cursos en línea a su ritmo. Métodos pasivos (observar) o totalmente autónomos (prueba y error sin guía) quedan relegados o se usan como complemento.

## 3. Análisis de Relaciones entre Variables (Tablas de contingencia y gráficos cruzados)

En este apartado, analizaremos cómo se **relacionan ciertas variables categóricas entre sí**. Usaremos tablas de contingencia (cruces de frecuencias) y gráficos de barras apiladas para visualizar cómo, por ejemplo, la preferencia de ambiente laboral varía según el género, o si la decisión de estudiar afuera se asocia al país del encuestado, etc. Esto nos ayudará a identificar patrones o diferencias significativas entre subgrupos.

### Género vs. Factor de Influencia en Aspiraciones

¿Hombres y mujeres difieren en qué factor dicen que más influye sus aspiraciones? Cruzaremos **género** con **factor de influencia**:

```python
cross_gen_factor = pd.crosstab(df['Your Gender'], df['Which of the below factors influence the most about your career aspirations ?'])
print(cross_gen_factor)

# Graficar barras apiladas
cross_gen_factor.T.plot(kind='bar', stacked=True, figsize=(8,5), color=["#66b3ff","#ff9999"])
plt.title("Influencia en aspiraciones por Género")
plt.ylabel("Número de encuestados")
plt.xlabel("Factor de influencia principal")
plt.xticks(rotation=45, ha='right')
plt.legend(title="Gender", loc="upper right")
plt.show()
```

**Observación:** Al analizar la tabla de contingencia y el gráfico apilado, notamos algunas diferencias interesantes por género:
- Tanto hombres como mujeres tienden a citar en alta proporción a **“Personas que cambiaron el mundo”** como una gran influencia, pero **los hombres lo hacen ligeramente más**. Es posible que los hombres Gen Z se sientan más inspirados por figuras históricas o líderes que produjeron cambios globales, en comparación con las mujeres.
- **“Mis padres”** es un factor muy mencionado en ambos géneros, pero parece ser **más predominante entre las mujeres**. Una porción ligeramente mayor de encuestadas femeninas eligió a sus padres como principal influencia comparado con los hombres. Esto sugiere que las mujeres Gen Z podrían apoyarse un poco más en la orientación familiar al pensar en su carrera.
- **“Redes sociales (LinkedIn)”** e **“Influencers de carreras exitosas”** muestran tasas similares entre géneros, quizás con leve aumento entre hombres en el caso de LinkedIn. En general, no hay una brecha enorme: ambos géneros están expuestos y motivados por contenido de carrera en redes, aunque podríamos conjeturar que hombres aprovechan un poco más LinkedIn en esta muestra.
- **“Personas de mi círculo (no familia)”** tuvo pocos votos en general, sin una diferencia clara de género.

En resumen, **hombres y mujeres comparten los principales motores de inspiración**, pero **las mujeres confían ligeramente más en la familia**, mientras que **los hombres algo más en héroes públicos o medios profesionales**. No obstante, las diferencias no son drásticas; ambos géneros valoran tanto ejemplos a seguir globales como el apoyo de sus padres.

### Género vs. Ambiente de Trabajo Preferido

Veamos si la preferencia por trabajo remoto/híbrido vs oficina varía entre hombres y mujeres:

```python
cross_gen_env = pd.crosstab(df['Your Gender'], df['What is the most preferred working environment for you.'])
print(cross_gen_env)

cross_gen_env.T.plot(kind='bar', stacked=True, figsize=(7,5), color=["#66b3ff","#ff9999"])
plt.title("Ambiente laboral preferido por Género")
plt.ylabel("Número de encuestados")
plt.xlabel("Ambiente de trabajo")
plt.xticks(rotation=0)
plt.legend(title="Gender")
plt.show()
```

**Observación:** Al cruzar **género con ambiente preferido**, vemos que **tanto hombres como mujeres presentan patrones muy similares** de preferencias. Ambos eligen mayoritariamente el esquema **híbrido** como top 1, seguido de las opciones remotas:

- En las opciones **remotas**, no se aprecia una diferencia notable: una fracción comparable de hombres y mujeres opta por “remoto con posibilidad de viajar” y “remoto 100%”. Quizá marginalmente más mujeres que hombres escogieron *remoto con viajes*, pero la diferencia es muy pequeña.
- **Oficina diaria** es impopular en ambos grupos, con números bajos tanto para hombres como para mujeres (virtualmente ninguno en el caso de las mujeres, y muy pocos hombres).

Así, **no hay una variación significativa por género en la preferencia de modalidad laboral**. Podemos concluir que **la preferencia por trabajos flexibles es transversal a género** en la Gen Z: mujeres y hombres coinciden en querer esquemas híbridos o remotos, confirmando que es una generación unida en esa expectativa laboral.

## 4. Evaluación de Correlaciones

Dado que varias columnas son categóricas, la correlación convencional de Pearson no aplica directamente. Sin embargo, podemos **codificar numéricamente algunas variables categóricas** (especialmente las dicotómicas o ordinales) para calcular correlaciones y visualizar posibles asociaciones fuertes entre variables.

Procederemos a crear una versión codificada del dataset para análisis de correlación:
- Variables **Sí/No** se mapearán a 1/0 (por ejemplo, *Would you work for a company with mission not clear*: Yes=1, No=0).
- Variables con **respuestas ordinales** (como la de 3 años con un empleador, o la de posgrado fuera de India, que tiene un orden implícito de compromiso) las mapearemos a números.
- **Género** mapearemos Male=1, Female=0 (por ejemplo) para ver si tiene alguna correlación con otras variables.
- La **escala 1-10 de impacto social** ya es numérica.
- Ignoraremos columnas no ordinales puramente nominales como “factor de influencia” o “carrera aspiracional” porque asignarles números arbitrarios no tiene sentido para correlación lineal. Solo incluiremos las que tengan una interpretación numérica o booleana.

Vamos a hacer esta codificación y calcular la matriz de correlación:

```python
# Copiar dataframe para codificación
df_enc = df.copy()

# Codificar género: Male=1, Female=0
df_enc['Your Gender'] = df_enc['Your Gender'].map({'Male': 1, 'Female': 0})

# Codificar posgrado fuera (ordinal):
# "Yes, I will earn and do that" -> 2, "No, but if someone could bear the cost I will" -> 1, "No I would not..." -> 0
edu_map = {
    'Yes, I will earn and do that': 2,
    'No, But if someone could bare the cost I will': 1,
    'No I would not be pursuing Higher Education outside of India': 0
}
df_enc['HigherEd_abroad'] = df_enc['Would you definitely pursue a Higher Education / Post Graduation outside of India ? If only you have to self sponsor it.'].map(edu_map)

# Codificar lealtad 3 años (ordinal): "No way..."->0, "hard but if right co..."->1, "Will work 3+ years"->2
loyal_map = {
    'No way, 3 years with one employer is crazy': 0,
    'This will be hard to do, but if it is the right company I would try': 1,
    'Will work for 3 years or more': 2
}
df_enc['Loyalty3yrs'] = df_enc['How likely is that you will work for one employer for 3 years or more ?'].map(loyal_map)

# Codificar misión no clara (Yes=1, No=0)
df_enc['Work_noMissionClear'] = df_enc['Would you work for a company whose mission is not clearly defined and publicly posted.'].map({'Yes': 1, 'No': 0})

# Codificar misión desalineada (Will work=1, Will NOT work=0)
df_enc['Work_misalignedMission'] = df_enc['How likely would you work for a company whose mission is misaligned with their public actions or even their product ?'].map({'Will work for them': 1, 'Will NOT work for them': 0})

# La columna de impacto social ya es numérica 1-10, podemos renombrarla para claridad
df_enc['Work_noSocialImpact_score'] = df_enc['How likely would you work for a company whose mission is not bringing social impact ?']

# Codificar ambiente laboral preferido (ordenándolo de más tradicional a más remoto):
env_map = {
    'Every Day Office Environment': 0,
    'Hybrid Working Environment with less than 15 days a month at office': 1,
    'Fully Remote with Options to travel as and when needed': 2,
    'Fully Remote with No option to visit offices': 3
}
df_enc['Preferred_env_code'] = df_enc['What is the most preferred working environment for you.'].map(env_map)

# Seleccionar solo las columnas numéricas codificadas relevantes
corr_vars = ['Your Gender', 'HigherEd_abroad', 'Loyalty3yrs', 'Work_noMissionClear', 'Work_misalignedMission', 'Work_noSocialImpact_score', 'Preferred_env_code']
corr_matrix = df_enc[corr_vars].corr()
print(corr_matrix)

# Graficar heatmap de correlación
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='RdBu', center=0)
plt.title("Matriz de correlación entre variables codificadas")
plt.show()
```

# Limpieza y transformación de datos

En esta sección, realizaremos la limpieza y transformación de los datos:
- Eliminaremos columnas irrelevantes (como el código postal).
- Normalizaremos valores de texto (eliminando espacios extra, corrigiendo errores tipográficos).
- Convertiremos respuestas binarias "Yes"/"No" a valores 1/0.
- Expandiremos columnas de selección múltiple en variables dummy (one-hot encoding de cada opción seleccionada).
- Aplicaremos codificación One-Hot a variables categóricas nominales.
- Aplicaremos Label Encoding a variables ordinales (por ejemplo: intención de posgrado, lealtad laboral, etc.).

```python
# 2.1 Eliminar columnas irrelevantes
if 'Your Current Zip Code / Pin Code' in df.columns:
    df.drop('Your Current Zip Code / Pin Code', axis=1, inplace=True)

# 2.2 Renombrar columnas para facilitar la manipulación (opcional pero útil)
df.rename(columns={
    'Your Current Country.': 'Country',
    'Your Gender': 'Gender',
    'Which of the below factors influence the most about your career aspirations ?': 'InfluenceFactor',
    'Would you definitely pursue a Higher Education / Post Graduation outside of India ? If only you have to self sponsor it.': 'PostGradAbroad',
    'How likely is that you will work for one employer for 3 years or more ?': 'Loyalty3Years',
    'Would you work for a company whose mission is not clearly defined and publicly posted.': 'WorkUndefinedMission',
    'How likely would you work for a company whose mission is misaligned with their public actions or even their product ?': 'WorkMisalignedMission',
    'How likely would you work for a company whose mission is not bringing social impact ?': 'WorkNoSocialImpact',
    'What is the most preferred working environment for you.': 'PreferredWorkEnv',
    'Which of the below Employers would you work with.': 'PreferredEmployer',
    'Which type of learning environment that you are most likely to work in ?': 'PreferredLearning',
    'Which of the below careers looks close to your Aspirational job ?': 'TargetCareer',
    'What type of Manager would you work without looking into your watch ?': 'PreferredManager',
    'Which of the following setup you would like to work ?': 'TeamSetup'
}, inplace=True)

# 2.3 Normalizar texto: eliminar espacios extras y corregir errores tipográficos comunes
# Eliminar espacios en extremos de cada valor de tipo string
df_obj = df.select_dtypes(include='object')
df[df_obj.columns] = df_obj.apply(lambda x: x.str.strip())

# Corregir errores ortográficos conocidos (ejemplos: "bare the cost" -> "bear the cost", "an learning environment" -> "a learning environment")
df.replace({"bare the cost": "bear the cost",
            "an learning environment": "a learning environment"}, regex=True, inplace=True)

# 2.4 Convertir respuestas binarias "Yes"/"No" a 1/0
# Para la pregunta de posgrado:
df['PostGradAbroad'] = df['PostGradAbroad'].map({
    'Yes, I will earn and do that': 1,
    'No, But if someone could bear the cost I will': 0
})
# Pregunta sobre misión no definida (Sí/No):
df['WorkUndefinedMission'] = df['WorkUndefinedMission'].map({'Yes': 1, 'No': 0})
# Pregunta sobre misión mal alineada (Will work/Will NOT work):
df['WorkMisalignedMission'] = df['WorkMisalignedMission'].map({
    'Will work for them': 1,
    'Will NOT work for them': 0
})

# 2.5 Expandir columnas de selección múltiple en variables dummy
# Columnas multi-selección identificadas: 'PreferredLearning' y 'TeamSetup'
# Usamos get_dummies con separador ', ' para obtener una columna por cada opción seleccionada.
learning_dummies = df['PreferredLearning'].str.get_dummies(sep=', ')
setup_dummies = df['TeamSetup'].str.get_dummies(sep=', ')

# Añadir al DataFrame principal
df = pd.concat([df, learning_dummies, setup_dummies], axis=1)
# Eliminar columnas originales de selección múltiple
df.drop(['PreferredLearning', 'TeamSetup'], axis=1, inplace=True)

# 2.6 One-Hot Encoding para variables categóricas nominales
nominal_cols = ['Country', 'Gender', 'InfluenceFactor', 'PreferredWorkEnv', 'PreferredEmployer', 'PreferredManager']
df = pd.get_dummies(df, columns=nominal_cols, drop_first=False)

# 2.7 Label Encoding para variables ordinales
# La columna Loyalty3Years tiene respuestas ordinales ("This will be hard to do, but if..." < "Will work for 3 years or more")
loyalty_map = {
    "This will be hard to do, but if it is the right company I would try": 0,
    "Will work for 3 years or more": 1
}
df['Loyalty3Years'] = df['Loyalty3Years'].map(loyalty_map)

# Asegurar que la columna WorkNoSocialImpact (impacto social) sea numérica (ordinal)
# En el dataset, esta columna es una escala numérica (ej. 1-10). Convertimos a int por seguridad.
df['WorkNoSocialImpact'] = pd.to_numeric(df['WorkNoSocialImpact'], errors='coerce')

# Verificamos que no queden columnas categóricas sin codificar (todas deben ser numéricas ahora, excepto el target)
print("Tipos de datos tras limpieza y codificación:\n", df.dtypes)
```

## 3. Codificación de la variable objetivo

Usaremos la columna **TargetCareer** (`Which of the below careers looks close to your Aspirational job ?`) como variable objetivo. Aplicaremos Label Encoding a esta variable para convertir las categorías de carreras aspiracionales en etiquetas numéricas.

```python
from sklearn.preprocessing import LabelEncoder

# Separar la variable objetivo
y = df['TargetCareer'].astype(str)  # convertir a str por seguridad si tiene tipo object
X = df.drop('TargetCareer', axis=1)

# Aplicar Label Encoding a la variable objetivo
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# Opcional: mostrar el número de clases distintas en la variable objetivo
print("Número de clases de carrera aspiracional:", len(le.classes_))
print("Ejemplos de clases:", list(le.classes_)[:5])
```

## 4. División de datos en entrenamiento y prueba

Ahora separaremos el conjunto de datos en entrenamiento (70%) y prueba (30%). Usaremos muestreo estratificado (`stratify`) para mantener la distribución de las clases objetivo en ambos conjuntos. Fijamos `random_state=42` para reproducibilidad.

```python
from sklearn.model_selection import train_test_split

# Dividir en entrenamiento y prueba con estratificación
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.30, stratify=y_encoded, random_state=42
)

print("Tamaño de entrenamiento:", X_train.shape[0], "instancias")
print("Tamaño de prueba:", X_test.shape[0], "instancias")

# Verificar distribución estratificada de un par de clases en ambos conjuntos
print("Distribución de clases en entrenamiento:", pd.Series(y_train).value_counts(normalize=True)[:5])
print("Distribución de clases en prueba:", pd.Series(y_test).value_counts(normalize=True)[:5])
```

## 5. Entrenamiento del modelo Random Forest

Entrenaremos un modelo **RandomForestClassifier** con los parámetros especificados:
- `n_estimators=100`
- `max_depth=7`
- `random_state=42`
- `oob_score=True` (para obtener el puntaje de validación *out-of-bag* en el conjunto de entrenamiento)

Luego ajustaremos el modelo con los datos de entrenamiento.


```python
from sklearn.ensemble import RandomForestClassifier

# Crear el modelo Random Forest con hiperparámetros optimizados
model = RandomForestClassifier(n_estimators=100, max_depth=7, random_state=42, oob_score=True)
model.fit(X_train, y_train)

# Después del entrenamiento, podemos obtener el score OOB y la precisión en entrenamiento
print(f"OOB Score (Out-of-Bag): {model.oob_score_:.3f}")
print(f"Accuracy en entrenamiento: {model.score(X_train, y_train):.3f}")
```

## 6. Evaluación del modelo

En esta sección evaluamos el desempeño del modelo entrenado utilizando el conjunto de prueba:
- **Accuracy general** en los datos de prueba.
- **Matriz de confusión** para visualizar predicciones vs. valores reales por clase (representada con un heatmap).
- **Reporte de clasificación** con precisión, recall y F1-score para cada clase.
- **Importancia de variables** mediante los valores de *feature importance* del modelo Random Forest (graficamos las más importantes).
- **Reducción de dimensiones (PCA)** para visualizar la distribución de las clases en dos dimensiones principales.

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# 6.1 Calcular accuracy en el conjunto de prueba
y_pred = model.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print(f"Accuracy en conjunto de prueba: {acc:.3f}")

# 6.2 Matriz de confusión
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("Matriz de Confusión - Random Forest")
plt.ylabel("Clase verdadera")
plt.xlabel("Clase predicha")
plt.show()

# 6.3 Reporte de clasificación (precisión, recall, F1 por clase)
print("Reporte de clasificación:")
print(classification_report(y_test, y_pred))

# 6.4 Importancia de variables
import numpy as np

feature_importances = model.feature_importances_
features = X.columns

# Ordenar las características por importancia descendente
indices = np.argsort(feature_importances)[::-1]
top_n = 10  # número de características más importantes a mostrar
top_features = features[indices][:top_n]
top_importances = feature_importances[indices][:top_n]

plt.figure(figsize=(10,6))
sns.barplot(x=top_importances, y=top_features)
plt.title("Top 10 características más importantes")
plt.xlabel("Importancia")
plt.ylabel("Características")
plt.show()

# 6.5 Reducción de dimensionalidad PCA para visualización en 2D
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Imputar valores faltantes: reemplazamos NaN con 0 (alternativamente, puedes usar la mediana)
X_no_nan = X.fillna(0)

# Escalar las características antes de PCA
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_no_nan)

# Aplicar PCA para reducir a 2 componentes principales
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_scaled)

print("Varianza explicada por las 2 primeras componentes:", pca.explained_variance_ratio_)

# Crear un DataFrame con las componentes PCA y la clase para graficar
pca_df = pd.DataFrame({
    'PC1': X_pca[:,0],
    'PC2': X_pca[:,1],
    'Clase': y_encoded  # usamos la codificación numérica de la clase
})

plt.figure(figsize=(8,6))
sns.scatterplot(data=pca_df, x='PC1', y='PC2', hue='Clase', palette='viridis', legend='brief', alpha=0.7)
plt.title("Distribución de clases en el plano PCA")
plt.xlabel("Componente Principal 1")
plt.ylabel("Componente Principal 2")
plt.show()
```
